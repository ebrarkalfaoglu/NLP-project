```python
!pip install transformers accelerate bitsandbytes sentencepiece --quiet #needed libraries

from transformers import AutoModelForCausalLM, AutoTokenizer

model_name = "Qwen/Qwen2.5-Coder-1.5B-Instruct"  #name of the model

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype="auto",
    device_map="auto"
) #load the model from huggingface and for the best data type and automatically putting model on gpu we selected auto

tokenizer = AutoTokenizer.from_pretrained(model_name) #tokanizer for text to tokens
```

```python
prompt = "write a quick sort algorithm." #test prompt

messages = [
    {"role": "system", "content": "You are Qwen, created by Alibaba Cloud. You are a helpful assistant."},
    {"role": "user", "content": prompt}
] #message structure

text = tokenizer.apply_chat_template(  #tokenizeri çalıştırdım
    messages,
    tokenize=False,
    add_generation_prompt=True
) #converts message into the chat template

model_inputs = tokenizer([text], return_tensors="pt").to(model.device)
```

```python
generated_ids = model.generate(
    **model_inputs,
    max_new_tokens=512 #maks length of answer
)

generated_ids = [
    output_ids[len(input_ids):] #removing original input tokens
    for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
]

response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0] #convert back to text
print(response) #print final output
```
